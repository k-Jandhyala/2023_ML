```python
import math
from scipy import stats
from scipy.stats import norm


#Calculate the z test statistic and p-value
def significance_proportion_test (successes, n, p0):
  #successes are the number of units of unacceptable quality
  #n is the sample size
  #p0 is the null hypothesis probability
  p_hat = successes/n
  z = ((p_hat-p0)/math.sqrt((p0*(1-p0))/n))
  p = norm.sf(abs(z))
  
  return z, p

#Interpret the p-value and state a conclusion
def interpret_sig_test_results(alpha, pval):
  return 'Fail to reject H0' if(alpha < pval) else 'Reject H0'

print("Hypothesis: The stock trader is no better than flipping a coin")
z_val, p_val = significance_proportion_test(10, 10, 0.05)
print("""
The Z value is {}
The P value is {}""".format(z_val, p_val))
print("The response: ", interpret_sig_test_results(z_val, p_val))

print("""
The response tells us that the Stock Trade is a master trader because he correctly predicted the stocks all 10 days. The percentage of this happening is about 0.00097%.""")
print('\033[1m' + "TLDR: " + '\033[0m' + "The deal is not a scam and you should probobly pay for the service")

print("""

""")
print('\033[1m' + "My Analasys: " + '\033[0m' + """I reject the rejection. My analasys, even though, shows that he is a skilled stock trader, he could have manipulated the market or something like that. I would not pay for this 
person's service because it doesn't make intuitvive sense. I thought that the scammer bought out like 100-1000 stocks of a small buisness and manipulated the market buy selling and buiing the stock, forcing it to react in the way 
the scammer wants it to.
""")
print('\033[1m' + "The real scam: " + '\033[0m' + """The man on the 1st of the 10 days sends out a million emails detailing the stocks and flips a coin(every peosn has a diffrent stock outcome). For 1/2 the people, it would be 
correct and they would recive new email the next day. The people who got incorrect were never mailed again(they lost trust in the service already). This continues for 10 days and then some 1000 people are left who got all the 
predicitons right. They have trust in the service and are willing to pay for it, even if the price is $50 and only 900 people pay it the scammer made about 45K.""")
```

**Doctest**:

Doctest is a library that is included in Python to allow generation of testing scenarios (a.k.a. test
vectors) for your functions. This functionality is extremely useful and important for any engineer
or computer scientist developing code as it allows module level (or unit) testing of functions and
methods. The tests are output from the standard Python interpreter and cut/pasted into
“docstrings”. Here’s a reference link. To insert a doctest in your code, you’ll need to import
doctest. Then to insert a test use three double quotes to delimit the test (e.g. put 3 quotes
before and after the test code. You may insert comments using the standard comment hashtag
symbol (#) to document the tests you are running. Then use three greater-than symbols to
identify the test code (e.g. setting up variables and calling the function. Finally, list the expected output and run `doctest.testmod()`. A message will then be printed indicating the pass or fail
condition of the test. Here is an example:

```python
import doctest as d

def helloWorld():
  print("Hello World!!")
def helloStudent(student):
  print("Hello ", student)
def helloStudents(student1, student2):
  print("Hello ", student1, " and ", student2)
def helloTeacher(teacher):
  return "Hello", teacher, ", how are you"
def helloPrincipal(principal, name):
  return "Hello", principal, "Hi, ", name

helloWorld()
helloStudent('Jim')
helloStudents('Jim', 'Jane')
print(helloTeacher('Jim'))
response1 = helloPrincipal('Jim', 'Jane')
response2 = helloPrincipal('Jim', 'Jane')
print(response1)
print(response2)

"""
>>> helloStudent('Jim')
Hello Jim
>>> helloStudents('Jim', 'Jane')
Hello Jim and Jane
"""

d.testmod()
```

# **Assignment - Queues**

Create a queue-based application which continually (hint: while True:) prompts a user and
inputs options (integer values) to perform the following:


````python
from collections import deque
queue = deque()

while True:
  choice = input("""Please enter a choice that corrosponsing to the following: 
    1 - Add a task to the queue
    2 - Display the first task in the queue
    3 - Complete and remove the first task in the queue
    4 - Remove all tasks in the queue
    5 - End
    : """)
  
  print(" ")

  if(choice == '1'):
    queue.append(input("What task should I add to your list: "))
  elif(choice == '2'):
    if(len(queue) == 0):
      print("There is nothing in your list, you should add something")
    else:
      print("The first item in you list is: ", queue[0])
  elif(choice == '3'):
    if(len(queue) == 0):
      print("There is nothing in your list, you should add something")
    else:
      queue.popleft()
  elif(choice == '4'):
    queue.clear()
  elif(choice == '5'):
    print('Goodbye User')
    break

  print(" ")
````
